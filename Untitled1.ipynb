```python
n = int(input("Nhập số lượng phần tử: "))

tong_chan = 0
tong_le = 0

for i in range(n):
    so = int(input("Nhập 1 số: "))

    tong_chan += so * (1 - so % 2)
    tong_le   += so * (so % 2)

print("Tổng số chẵn:", tong_chan)
print("Tổng số lẻ:", tong_le)
```
